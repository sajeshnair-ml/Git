**Agenda : Git & Github**

**What is Git**

Git is a distributed version control system (VCS) used to track changes in source code during software development.

It helps you:
- Track code history
- Revert to previous versions
- Collaborate with multiple developers
- Manage branches safely

Key Features:
- Distributed system (each developer has a full repository copy)
- Fast and lightweight
- Strong branching and merging support
- Maintains commit history with author and timestamps
- Works offline


**What is GitHub**

GitHub is a cloud-based platform that hosts Git repositories online and enables collaboration.

It provides:
- Remote repository storage
- Pull Requests and Code Reviews
- Issue tracking
- Project boards
- CI/CD integrations
- Open-source collaboration

Simple understanding:
Git = Version Control Tool  
GitHub = Hosting and Collaboration Platform


**Why Use Git and GitHub**

- Enables team collaboration
- Provides cloud backup of code
- Helps in open-source contributions
- Acts as a professional portfolio
- Tracks progress in Data Science and ML projects
- Industry standard in software development


**Basic Git Workflow**

Working Directory → Staging Area → Local Repository → Remote Repository


Common Git Commands:

git init  
git clone <repo_url>  
git status  
git add .  
git commit -m "Your commit message"  
git branch  
git checkout <branch_name>  
git merge <branch_name>  
git pull  
git push  


**Core Concepts**

Repository – A project folder tracked by Git  
Commit – A snapshot of changes  
Branch – A separate line of development  
Merge – Combining changes from different branches  
Clone – Copying a remote repository locally  
Pull – Fetch and merge changes from remote  
Push – Uploading local commits to remote  
Fork – Copy of someone else's repository  


**Git vs GitHub**

Git:
- Version control system
- Works locally
- Command-line based
- Can work offline
- Tracks code changes

GitHub:
- Cloud hosting platform
- Works online
- Web-based interface
- Requires internet
- Manages collaboration


**Conclusion**

Git helps you manage and track code changes locally.  
GitHub helps you store, collaborate, and showcase projects online.

Mastering Git and GitHub is essential for Software Developers, Data Scientists, ML Engineers, and DevOps Engineers.
